```python
from pyspark.sql import SparkSession
from pyspark. sql. functions import from_json, col, struct, to_json
from pyspark.sql.types import StructType, StructField, IntegerType, DoubleType
```

```python
spark = SparkSession.builder \
    .appName("FraudDetection") \
    .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.3") \
    .getOrCreate()
spark.sparkContext.setLogLevel("WARN")
```

```python
# 1. Load Static User Data (CSV)
users_df = spark.read.csv("data/user_table.csv", header=True, inferSchema=True)
```

```python
# 2. Read Streaming Data from Kafka
tx_schema = StructType([
 StructField("tx_id", IntegerType(), True),
 StructField("userId", IntegerType(), True),
 StructField("amount", DoubleType(), True),
 StructField("timestamp", DoubleType(), True)
])
kafka_stream = spark.readStream \
 .format("kafka") \
 .option("kafka.bootstrap.servers", "kafka:9092") \
 .option("subscribe", "fraud-detection") \
 .load()
```

```python
# 3. Parse and Filter Data
parsed_stream = kafka_stream.select(from_json(col("value").cast("string"),
tx_schema).alias("tx")).select("tx.*")
fraud_stream = parsed_stream.filter(col("amount") > 10000.0)
```

```python
# 3(b). Parse and Filter Data
parsed_stream_high = kafka_stream.select(from_json(col("value").cast("string"),
tx_schema).alias("tx")).select("tx.*")
fraud_stream_high = parsed_stream.filter(col("amount") > 5000.0)
```

```python
# 4. Enrich Stream with User Details
enriched_fraud = fraud_stream.join(users_df, "userId")
```

```python
# 4(b). Enrich Stream with User Details
enriched_fraud_high = fraud_stream_high.join(users_df, "userId")
```

```python
# 5. Format for output Kafka topic
output_stream = enriched_fraud \
 .withColumn("value", to_json(struct("*")).cast("string")) \
 .select("value")
```

```python
# 5(b). Format for output Kafka topic
output_stream_high = enriched_fraud_high \
 .withColumn("value", to_json(struct("*")).cast("string")) \
 .select("value")
```

```python
# 6. Write Stream to 'fraud-notification' Topic
query = output_stream.writeStream \
 .format("kafka") \
 .option("kafka.bootstrap.servers", "kafka:9092") \
 .option("topic", "fraud-notification") \
 .option("checkpointLocation", "/workspace/checkpoints") \
 .start()
```

```python
# 6(b). Write Stream to 'fraud-notification' Topic
query_high = output_stream_high.writeStream \
 .format("kafka") \
 .option("kafka.bootstrap.servers", "kafka:9092") \
 .option("topic", "high-alert") \
 .option("checkpointLocation", "/workspace/checkpoints_high_alerts") \
 .start()
query.awaitTermination()
query_high.awaitTermination()
```
